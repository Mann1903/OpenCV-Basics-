# MORPHOLOGICAL OPERATIONS



```python
#erosion
import cv2
import numpy as np
img=cv2.imread(r"C:\Users\admin\Downloads\j.png")
kernel=np.ones((5,5),np.uint8)
erosion=cv2.erode(img,kernel,iterations=1)
cv2.imshow("image",img)
cv2.imshow("eroded image",erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#dilation
import cv2
import numpy as np
img=cv2.imread(r"C:\Users\admin\Downloads\j.png")
kernel=np.ones((5,5),np.uint8)
dilate=cv2.dilate(img,kernel,iterations=1)
cv2.imshow("image",img)
cv2.imshow("daileted image",dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Harris corner detection
```

```python
import cv2
import numpy as np
  
# path to input image specified and 
image = cv2.imread(r"C:\Users\admin\Downloads\stop-sign1.png")
  
# grayscale color space
operatedImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  
# modify the data type
operatedImage = np.float32(operatedImage)
  
# apply the cv2.cornerHarris method
dest = cv2.cornerHarris(operatedImage, 2, 5, 0.07)
  
# Results are marked through the dilated corners
dest = cv2.dilate(dest, None)
  
# Reverting back to the original image,
# with optimal threshold value
image[dest > 0.01 * dest.max()]=[0, 0, 255]
  
# the window showing output image with corners
cv2.imshow('Image with Borders', image)
  
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
import cv2
 
# Loading the image
img = cv2.imread(r"C:\Users\admin\Downloads\stop-sign1.png")
 
 # Converting image to grayscale
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
 
# Applying SIFT detector
sift = cv2.SIFT_create()
kp = sift.detect(gray, None)
 
# Marking the keypoint on the image using circles
img=cv2.drawKeypoints(gray,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
 
cv2.imwrite('image-with-keypoints.jpg', img)
cv2.imshow('SIFT_image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
